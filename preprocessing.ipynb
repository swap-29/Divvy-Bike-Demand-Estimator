```python
# import the necessary libraries
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

# Dataset 1: Divvy Bike Data

```python
# Read the various csv files from Nov 2019 to Oct 2020 (includes Oct 2019 as well)
d1 = pd.read_csv("divvy/Divvy_Trips_2019_Q4.csv")
d2 = pd.read_csv("divvy/divvydata_Jan-Mar2020.csv")
d3 = pd.read_csv("divvy/202004-divvy-tripdata.csv")
d4 = pd.read_csv("divvy/202005-divvy-tripdata.csv")
d5 = pd.read_csv("divvy/202006-divvy-tripdata.csv")
d6 = pd.read_csv("divvy/202007-divvy-tripdata.csv")
d7 = pd.read_csv("divvy/202008-divvy-tripdata.csv")
d8 = pd.read_csv("divvy/202009-divvy-tripdata.csv")
d9 = pd.read_csv("divvy/202010-divvy-tripdata.csv")
```

```python
# since the data of the 4th quarter of 2019 has different attributes than the rest of the other datasets, rename the attributes 
d1.rename({'trip_id': 'ride_id', 'start_time':'started_at', 'end_time': 'ended_at', 'from_station_name':'start_station_name', 'from_station_id': 'start_station_id', 'to_station_name': 'end_station_name', 'to_station_id': 'end_station_id', 'usertype': 'member_casual'}, axis=1, inplace=True)
d1
```

```python
# extract the necessary columns from all the datasets
d1 = d1[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d2 = d2[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d3 = d3[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d4 = d4[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d5 = d5[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d6 = d6[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d7 = d7[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d8 = d8[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d9 = d9[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]

d3
```

```python
# merge all the datasets to get the complete data for the Divvy bikes
divvy_data = pd.concat([d1, d2, d3, d4, d5, d6, d7, d8, d9])
divvy_data
```

```python
# Convert the date columns (start and end times) in the divvy data to datetime format 
divvy_data['started_at'] = pd.to_datetime(divvy_data.started_at)
divvy_data['ended_at'] = pd.to_datetime(divvy_data.ended_at)

# make new columns 'Year', 'Month', 'Day' and 'Hour' based on the 'started_at' column
divvy_data['Year'] = divvy_data.started_at.dt.year
divvy_data['Month'] = divvy_data.started_at.dt.month
divvy_data['Day'] = divvy_data.started_at.dt.day
divvy_data['Hour'] = divvy_data.started_at.dt.hour

# save the resuting dataframe to a csv
divvy_data.to_csv("divvy_final.csv", index=False)
divvy_data
```

# Dataset 2 - Weather Data

```python
# read the weather data from the csv file into a pandas dataframe
weather = pd.read_csv("weather/weather.csv")
weather
```

```python
# extract the relevant columns needed for our predictions
weather = weather[['Date', 'Temperature', 'Humidity', 'Wind Speed', 'Wind Gust', 'Precipitation', 'Condition']]
weather
```

```python
# convert the weather date to datetime format
weather['Date'] = pd.to_datetime(weather.Date)

# create new columns 'Year', 'Month', 'Day' and 'Hour' based on the date column
weather['Year'] = weather.Date.dt.year
weather['Month'] = weather.Date.dt.month
weather['Day'] = weather.Date.dt.day
weather['Hour'] = weather.Date.dt.hour

# save the dataframe to a csv file
weather.to_csv("weather_final.csv", index=False)
weather
```

```python
# merge both the divvy and weather data based on the extracted year, month, day and hour
merged_df = divvy_data.merge(weather, how='inner', on=['Year', 'Month', 'Day', 'Hour'])

# drop the 'Date' column from the merged data , since we are using the 'started_at' column from divvy_data
merged_df.drop(['Date'], axis=1, inplace=True)
merged_df
```

# Dataset 3: Holidays

```python
# read the list of holidays from the csv file
holidays = pd.read_csv("holidays/list-us-public-holidays-454j.csv", encoding="ISO-8859-1")
holidays
```

```python
# remove the last 3 rows, since they are all NaNs
holidays.drop(holidays.tail(3).index,inplace=True)
```

```python
holidays
```

```python
# extract the relevant columns
holidays = holidays[['Unnamed: 1', 'Official Name', 'Remarks']]

# rename the Unnamed column to Date
holidays.rename({'Unnamed: 1': 'Date'}, axis=1, inplace=True)
holidays
```

```python
# convert the date column to datetime format (for easier processing)
holidays.Date = pd.to_datetime(holidays.Date)
```

```python
# get the list of dates for spring break
vacations = {'Spring Break': pd.date_range('4/6/2020', '4/10/2020')}

for desc, hols in vacations.items():
    for hol in hols:
        # get the date from the timestamp
        hol=pd.to_datetime(hol).date()
        # and append to the dataframe
        holidays.loc[len(holidays.index)] = [hol, desc, '']
        
# assign it to the Date column
holidays.Date = pd.to_datetime(holidays.Date)
holidays
```

```python
# get list of dates for summer vacation
d = pd.date_range('6/18/2020', '9/3/2020') # datetimeindex obj
d1 = list(holidays['Date']) # series
# reassign the dates from the Date column in d1
d1=[x.date() for x in d1]
d2 = []
for date in d:
    # if the date isn't already in d1, add to the empty list d2
    if date.date() not in d1:
        d2.append(date.date())
        
# and assign it to a new dataframe
sdates = pd.DataFrame(d2, columns=['Date'])
# give the value as Summer Vacation
sdates['Official Name'] = 'Summer Vacation'
# convert the date into datetime format
sdates['Date'] = pd.to_datetime(sdates.Date)
sdates
```

```python
# get list of dates for winter vacation
d = pd.date_range('12/23/2019', '1/3/2020') 
d1 = list(holidays['Date']) # series

# reassign the dates in the Date column
d1=[x.date() for x in d1]
d2 = []
for date in d:
    # if the date isn't in d1
    if date.date() not in d1:
        # add to the empty list d2
        d2.append(date.date())
# create a new dataframe with that list
wdates = pd.DataFrame(d2, columns=['Date'])
# and assign a value of Winter Vacation for all the records
wdates['Official Name'] = 'Winter Vacation'
# convert the Date attribute to datetime format
wdates['Date'] = pd.to_datetime(wdates.Date)
wdates
```

```python
# merge the holidays and winter break dates
hols = holidays.merge(wdates, how='outer', on=['Date'])
# sort by Date in ascending order
hols.sort_values(by='Date', inplace=True)
hols
```

```python
# create a new column Name and assgn it h Official Name_x, if NaN, assign with Official Name_y
hols['Name']=hols['Official Name_x'].fillna(hols['Official Name_y'])
hols
```

```python
# remove the unecessary columns from the dataframe
hols.drop('Official Name_y', axis=1, inplace=True)
hols.drop('Remarks', axis=1, inplace=True)
hols.drop('Official Name_x', axis=1, inplace=True)
hols
```

```python
# merge the merged dataframe with the summer dates
hols = hols.merge(sdates, how='outer', on=['Date'])
# sort by Date in ascending order
hols.sort_values(by='Date', inplace=True)
hols
```

```python
# assign the Name column with values from the Name column, if NaN, fill with values from Official Name
hols['Name']=hols['Name'].fillna(hols['Official Name'])
hols
```

```python
# drop Official Name from the dataframe
hols.drop('Official Name', axis=1, inplace=True)
```

```python
# save the results into a csv file
hols.to_csv('hols.csv', index=False)
```

```python
# get the list of dates from 11/1/2019 to 11/30/2019
d = pd.date_range('11/1/2019', '11/30/2020')
d1 = list(hols['Date'])
# reassign dates for the Date column
d1 = [x.date() for x in d1]
d3 = list(wdates['Date']) 
# reassign dates for the Date column in d3
d3=[x.date() for x in d3]
d2 = []
for date in d:
    # if the date is not in the merged data as well as not in the winter dates
    if date.date() not in d1 and date.date() not in d3:
        # add the date to the empty list d2
        d2.append(date.date())
# create a dataframe with d2 as the Date column
dates = pd.DataFrame(d2, columns=['Date'])
# if no name, assign as Not Holiday
dates['Official Name'] = 'Not Holiday'
# convert the Date attribute to datetime format
dates['Date'] = pd.to_datetime(dates.Date)
dates
```

```python
dates.Date = pd.to_datetime(dates.Date)

# check if the day is a weekend, if so, rerwite its value to Weekend
dates['Official Name']=np.where(dates.Date.dt.weekday > 4 , 'Weekend', 'Not Holiday')
dates
```

```python
hols.Date = pd.to_datetime(hols.Date)
dates.Date = pd.to_datetime(dates.Date)

# merge the previously merged dataset with the dates dataset
hol = hols.merge(dates, how='outer', on=['Date'])
# and sort by Date in ascending order
hol.sort_values(by='Date', inplace=True)
hol
```

```python
# create a new column Holiday_Name and assign Official Name values to that column, if NaN, assign the corresponding values from the Name column
hol['Holiday_Name']=hol['Official Name'].fillna(hol['Name'])
```

```python
# drop the unnecessary columns
hol.drop('Official Name', axis=1, inplace=True)
hol.drop('Name', axis=1, inplace=True)
```

```python
hol
```

```python
# extract the Year, Month and Day columns and create them as separate columns
hol['Year'] = hol.Date.dt.year
hol['Month'] = hol.Date.dt.month
hol['Day'] = hol.Date.dt.day
hol
```

```python
# check for weekend in the final dataset
hol['Weekend'] = np.where(hol.Date.dt.weekday > 4, 1, 0)
# reset the index to start from 0
hol.reset_index(inplace=True)
hol
```

```python
# prevents overwriting of existing holidays
for i in range(len(hol['Date'])):
    if hol['Weekend'][i] == 1:
        if hol['Holiday_Name'][i] == "Not Holiday":
            hol['Holiday_Name'][i] = 'Weekend'
hol
```

```python
# save the dataset to a csv
hol.to_csv('holidays_final.csv', index=False)
```

```python
# merge the final holidays dataset with the divvy + weather data
final_merged_data = merged_df.merge(hol, how='inner', on=['Year', 'Month', 'Day'])
# and drop the Date column
final_merged_data.drop("Date", axis=1, inplace=True)
final_merged_data
```

```python
# save the final resulting dataframe to a csv (serves as the unclean merged data)
final_merged_data.to_csv("final_merged_data.csv", index=False)
```

```python
# drop the Weekend attribute
final_merged_data.drop("Weekend", axis=1, inplace=True)
```
