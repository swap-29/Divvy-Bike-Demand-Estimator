```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
holidays = pd.read_csv("list-us-public-holidays-454j.csv", encoding="ISO-8859-1")
holidays
```

```python
holidays.drop(holidays.tail(3).index,inplace=True)
```

```python
holidays = holidays[['Unnamed: 1', 'Official Name', 'Remarks']]
holidays.rename({'Unnamed: 1': 'Date'}, axis=1, inplace=True)
holidays
```

```python
holidays.Date = pd.to_datetime(holidays.Date)
print(type(holidays['Date'][0]))
"""# holidays.drop('Remarks', axis=1, inplace=True)
holidays['Year'] = holidays.Date.dt.year
holidays['Month'] = holidays.Date.dt.month
holidays['Day'] = holidays.Date.dt.day
holidays"""
holidays
holidays[holidays['Date'] == "2019-12-25"]
```

```python
vacations = {'Spring Break': pd.date_range('4/6/2020', '4/10/2020'), 'Summer Vacation': pd.date_range(start='6/18/2020', end='9/3/2020')}
# print(vacations)

# for date in pd.date_range('12/1/2019', '12/1/2020'):
#     print(type(date))
for desc, hols in vacations.items():
    for hol in hols:
        hol=pd.to_datetime(hol).date()
        holidays.loc[len(holidays.index)] = [hol, desc, '']
#holidays[holidays['Date'] == '12/25/2019']
#print(type(holidays['Date'][0]))
holidays.Date = pd.to_datetime(holidays.Date)
holidays
```

```python
d = pd.date_range('12/23/2019', '1/3/2020') # datetimeindex obj
d1 = holidays['Date'] # series
d1=list(d1)
d1=[x.date() for x in d1]
d2 = []
for date in d:
    if date.date() not in d1:
        d2.append(date.date())
wdates = pd.DataFrame(d2, columns=['Date'])
wdates['Official Name'] = 'Winter Vacation'
wdates['Date'] = pd.to_datetime(wdates.Date)
# dates['Year'] = dates.Date.dt.year
# dates['Month'] = dates.Date.dt.month
# dates['Day'] = dates.Date.dt.day
wdates
```

```python
hols = holidays.merge(wdates, how='outer', on=['Date'])
hols.sort_values(by='Date', inplace=True)
#hols.drop("Remarks", axis=1, inplace=True)
#hols[hols['Date']=='2019-12-25']
#hols[hols['Date']=='2019-12-24']
hols
```

```python
hols['Name']=hols['Official Name_x'].fillna(hols['Official Name_y'])
hols
```

```python
hols.drop('Official Name_y', axis=1, inplace=True)
hols.drop('Remarks', axis=1, inplace=True)
hols.drop('Official Name_x', axis=1, inplace=True)
hols
```
